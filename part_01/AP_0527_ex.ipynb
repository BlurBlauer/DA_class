**[연습문제 41]** `성별` 컬럼의 값을 `남`, `여`로 바꾸시오.

```python
cs['성별'] = np.where(cs['성별'] == 1, '남', '여') # '성별' 칼럼이 1일경우 '남', 그렇지 않다면 '여'로 변경하여 다시 '성별' 칼럼에 저장.
```

```python
cs['성별'] = cs['성별'].apply(lambda x: '남' if x == 1 else '여')
```

```python
cs.query('성별 == "여"')
```

**[연습문제 42]** `구매일자`에서 `요일`을 추출하여 새로운 컬럼 `판매요일`을 만드시오.

```python
tr.info() #데이터 프레임의 형식을 확인 --> 요일 데이터를 정수 형태로 변환해야 연산이 가능함.
```

```python
tr['판매요일'] = tr.구매일자.astype(str).astype('datetime64').dt.day_name()
#날짜 형태로 변환하여 요일을 추출. 그리고 이를 판매요일 칼럼에 저장.
#한글로 저장하고 싶다면 apply() 함수를 사용해야함.
```

**[연습문제 43]** `거주지역`을 시와 구 또는 도와 시로 나누어 두 개의 새로운 컬럼(`거주지역_광역`, `거주지역_기초`)을 만드시오.

```python
cs['거주지역_광역'] = cs.거주지역.apply(lambda x: x.split()[0])
#split() 함수로 문자열을 분리. [0]로 인덱싱하여 시 이름만을 인덱싱
cs['거주지역_기초'] = cs.거주지역.apply(lambda x: x.split()[1])

cs.head()
# 난이도가 있는 문제. 미리 복습하여 시험에 대비할 것.
```

**[연습문제 44]** `60대 고객`의 `총 구매건수`는 얼마인가?

```python
# 나이 정보는 cs에 구매정보는 tr에 있기 때문에 먼저, 두 데이터프레임을 merge해야 한다.
md = pd.merge(cs, tr, on='ID')
md.head()
```

```python
md.query('60 <= 연령 <= 69').shape[0] #10만건 중 60대 고객이 구매한 건수.
```

**[연습문제 45]** `남녀별` `총구매액`, `구매건수`, `평균구매액`, `최대구매액`, `최소구매액`을 계산하여 출력하시오. 

```python
md.groupby('성별').agg({'구매금액' : [('총구매액', 'sum'), ('구매건수', 'size'),
                                ('평균구매액', 'mean'),('최대구매액', 'max'), ('최소구매액', 'min')]}).reset_index()
#agg()함수를 통해 각각의 연산을 한꺼번에 처리 가능.
```

**[연습문제 46]** `총 구매액`이 가장 많은 사람부터 적은 사람 순으로 정렬하여 `상위 10명`만 출력하시오. 

```python
md.groupby(['ID'])['구매금액'].sum().index[:10] #상위에 있는 10명의 ID를 출력.
```

```python
md.groupby(['ID'])['구매금액'].sum().sort_values(ascending = False)[:10]
```

**[연습문제 47]** `서울`에 사는 고객들의 `지역별(광역)` `구매액`를 계산하여 출력하시오. 단, 지역명은 열로 나타낼 것.

```python
md.query("거주지역_광역 == '서울'").pivot_table(values= '구매금액', columns='거주지역_기초', 
               aggfunc = sum)
#'서울에 거주하는 사람들' 이라는 조건을 query를 통해 추출한 뒤, pivot_table을 통해서 '거주지역_기초'(칼럼)별로 구매금액의 합계를 계산. 
```

**[연습문제 48]** 각 `고객` 별로 `월` 별 `판매액(amount)`의 합계를 계산하여 출력하시오.

```python
md['구매일자'] = md['구매일자'].astype('str')
md["구매 월"] = md['구매일자'].str[4:6] 
md.head()
#구매일자를 숫자열에서 문자열로 바꾼뒤에 슬라이싱하여 '구매 월'이라는 칼럼을 생성함.
```

```python
#방법 1 - groupby를 통해서 ID와 구매 월에 따른 구매금액의 합계를 게산.
md.groupby(['ID','구매 월'])['구매금액'].sum()
```

```python
#방법 2 - pivot_table 기능을 통해서  ID, 구매 월 에 따른 구매금액의 합계를 계산, 값이 없는 월에 대해서는 0을 채움.
md.pivot_table(values= '구매금액', index = 'ID', columns ='구매 월', 
               aggfunc = sum, fill_value = 0 )
```
